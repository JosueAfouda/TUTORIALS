```python
# Importation de l'interface pyplot de Matplotlib

import matplotlib.pyplot as plt
```

# Trac√©s lin√©aires

```python
# Trac√© d'une courbe (line plot)
annee = [1950, 1970, 1990, 2010]
population = [2600000, 3700000, 5300000, 7000000]

plt.plot(annee, population)
plt.xlabel("Ann√©e")
plt.ylabel("Population (millions)")
plt.title("Evolution de la population")
plt.show()
```

```python
# On peut personnaliser le style de ligne, le marqueur et la couleur
plt.plot(annee, population, linestyle='--', marker='v', color='r')
plt.xlabel("Ann√©e")
plt.ylabel("Population (millions)")
plt.title("Evolution de la population")
plt.show()
```

```python
# Importez le fichier 'gapminder.txt'

gap = pd.read_csv("gapminder.txt")

# Affichez les cinq premi√®res lignes de gap

gap.head()
```

```python
# Tracez un nuage de points montrant l'esp√©rance de vie (life_exp) en fonction du PIB (gdp_cap)

plt.figure(figsize = (12, 7))

plt.scatter(gap['gdp_cap'], gap['life_exp'])

plt.xlabel("PIB")

plt.ylabel("Esp√©rance de vie)")

plt.title("Esp√©rance de vie VS PIB")

plt.show()
```

```python
# Reprenez le code du graphique pr√©c√©dent en ajoutant plt.xscale('log') et plt.grid(True)

plt.figure(figsize = (12, 7))

plt.scatter(gap['gdp_cap'], gap['life_exp'])

plt.xscale('log')

plt.xlabel("Log du PIB")

plt.ylabel("Esp√©rance de vie)")

plt.title("Esp√©rance de vie VS PIB")

plt.grid(True)

plt.show()
```

```python
gap.cont.unique()
```

```python
# Tracez un nuage de points montrant l'esp√©rance de vie en fonction du PIB dans le continent africain

afr = gap[gap.cont == "Africa"]

plt.figure(figsize = (12, 7))

plt.scatter(afr['gdp_cap'], afr['life_exp'])

plt.xlabel("PIB")

plt.ylabel("Esp√©rance de vie)")

plt.title("Esp√©rance de vie VS PIB")

plt.show()
```

# Diagramme √† barres

```python
# Exemple de diagramme √† barre

x1 = [1, 3, 4, 5, 6, 7, 9]

y1 = [4, 7, 2, 4, 7, 8, 3]

plt.bar(x1, y1, label="Blue Bar", color='b')

plt.xlabel("nombre")

plt.ylabel("Taille")

plt.title("Exemple de diagramme √† barre")

plt.legend()

plt.show()
```

```python
# Cr√©ez une dataframe du PIB moyen par continent

pib_moy = gap.groupby("cont")['gdp_cap'].mean()
pib_moy = pd.DataFrame(pib_moy)
pib_moy
```

```python
# Construire un diagramme √† barres montrant le PIB moyen par continent

plt.bar(pib_moy.index, pib_moy['gdp_cap'], label = "PIB per continent", color = 'g')

plt.xlabel("Continent")

plt.ylabel("PIB moyen")

plt.title("PIB moyen par continent")

plt.legend()

plt.show()
```

# Diagramme circulaire

```python
# Exemple de diagramme circulaire
labels = "S1", "S2", "S3"
sections = [56, 66, 24]
plt.pie(sections, labels = labels, colors = ['c', 'g', 'y'])
plt.title("Exemple de graphique circulaire")
plt.show()
```

```python
# On peut customiser le graphique pr√©c√©dent

plt.pie(sections, labels = labels, colors = ['c', 'g', 'y'],
        explode = (0, 0.1, 0.1), # Essayez aussi explode = (0, 0.1, 0)
        autopct = '%1.2f%%')

plt.title('Exemple de diagramme circulaire') 

plt.show()
```

```python
# Construisez un diagramme circulaire montrant le PIB moyen par continent

plt.pie(pib_moy.gdp_cap, labels = pib_moy.index)
plt.title("PIB moyen par continent")
plt.show()
```

# Histogramme

```python
# Construction d'un histogramme

plt.hist(apple.Close)
plt.title("Histogramme du prix de cloture de APPLE")
plt.show()
```

```python
# Changement du nombre de bacs (bins)

plt.hist(apple.Close, bins = 20)
plt.title("Histogramme du prix de cloture de APPLE")
plt.show()
```

```python
# Avec une boucle for, construire les histogrammes des variables population, life_exp et gdp_cap

for i in ["population", "life_exp", "gdp_cap"]:
  plt.hist(gap[i])
  plt.title(i)
  plt.show()
```

# Subplots et Sauvegarde d'un graphique

```python
plt.subplot(2, 1, 1)
plt.hist(apple.Close)
plt.ylabel("Prix de cloture")
plt.title("Histogramme")

plt.subplot(2, 1, 2)
plt.boxplot(apple.Close, vert=False)
plt.yticks([1], ["Prix de cloture"])
plt.title("Boite √† moustache")

plt.tight_layout()
plt.savefig("AppleClosedPrice.png", dpi=900)
```

```python
noms_variables = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race",
    "sex", "capital-gain", "capital-loss", "hours-per-week",
    "native-country", "class"
]

df = pd.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",
    header=None,
    names = noms_variables
)

df.head()
```

```python
df.dtypes
```

```python
# Colonne num√©riques
cols_num = df.select_dtypes(exclude="object").columns.to_list()
cols_num
```

```python
plt.figure(figsize=(18,7)) 
for i,col in enumerate(cols_num):
    plt.subplot(2, len(cols_num), i+1)
    plt.hist(df[col])
    plt.title(col)
for i,col in enumerate(cols_num):
    plt.subplot(2, len(cols_num), i+len(cols_num)+1)
    plt.boxplot(df[col], vert=False)
    plt.yticks([])
plt.tight_layout()
plt.savefig('Visualisation.png', dpi=900)
```
